# I: Introduction and Research Problem
The United States confirmed its first case of COVID-19 on January 21, 2020 in Washington after a resident tested positive for the virus and, in an effort to slow the spread of the virus, the government announced a series of lockdowns and mandatory shutdowns of non-essential businesses in March (“A Timeline of COVID-19 Developments in 2020,” 2021; Bauer, Broady, Edelberg, & O’Donnell, 2022)

Consequently, these measures negatively affected the country's economy, and the US experienced its greatest recession yet after GDP fell by a 32.9% annualized rate in the second quarter of 2020, the deepest decline ever recorded since records began in 1947 (Mutikani, 2020). 

The US economy has seen recovery in 2021,however, the emergence of the Omicron variant threatens the economy and health care system of the country once again as the high transmissibility of the new variant caused record breaking infection numbers nationwide and put a strain on the health care system of the country. 

Early data from the first two weeks of January 2022 show that the US recorded an average of more than 800,000 new cases a day while hospitalizations of people with the virus also reached a new high; and with people using at-home test kits and with some infections going undetected and/or unreported, these numbers could be an undercount of actual infections in the country (“Covid in the U.S.: Latest Map and Case Count,” 2022). Furthermore, with the rise in hospitalizations among COVID-19 patients, hospitals are also grappling with administering proper care because of medical staff shortage and declining availability of ICU beds in hospitals as reports show that nineteen states in the country have less than 15% remaining capacity in their ICU units, with four of them having less than 10% (Caldwell & Hanna, 2022). 

Vaccination is not only necessary for us to prevent serious COVID-19 cases among the vulnerable and the overloading of the healthcare sector; it is necessary for us to achieve herd immunity and the coveted pre-COVID lifestyle. Herd immunity would be the state at which the capacity of the virus to spread is severely debilitated, resulting in a societal environment where the virus dies down naturally and the vulnerable who are unvaccinated are naturally protected; this is much like how modern society has conquered deadly illnesses like measles and smallpox. In addition to that, the lack of herd immunity and generally low vaccination rates allow the virus to continue to circulate and infect many individuals, which increases the possibility of the mutation of the COVID-19 virus (“The effects of virus variants on COVID-19 vaccines,” 2021). Given the arising COVID-19 variants with increased levels of infectiousness, assured herd immunity may require as much as 85% of the population to be fully vaccinated (Plans, Torner, Godoy, Jané, 2014). 

Unfortunately, herd immunity is out of reach, even for a vaccine-producing country like the United States. Vaccine hesitancy continues to run rampant in the United States, specially among certain demographic groups. Based on the COVID-19 vaccine monitoring data presented by the Kaiser Family Foundation last November, unvaccinated adults in the U.S tend to be younger, less educated, lean republican, religious, and non-white (Kirzinger et al., 2021). 

The following explain the possible reason/s of each demographic group's vaccine hesitancy:

***Racial and Ethnic Minority Groups***

Studies show that racial and ethnic minority groups tend to experience the impacts of the COVID-19 pandemic more profoundly than their counterparts and are at higher risk of being exposed to the virus despite having the least resources to seek proper medical care in case of infection (Parker, Minkin, & Bennett, 2020). Because of these, inoculation in these demographic groups is crucial in ensuring their safety; however, history of racial discrimination and inequities in the U.S health care system, such as histories of unconsented sterilization and unethical experimentations towards minorities, contribute to mistrust in the health care system and vaccine hesitancy among these demographic groups (“Health Equity Considerations and Racial and Ethnic Minority Groups,” 2021) In addition to that, language barriers and the lack of flexibility in their working conditions, that would enable them to take the time to get vaccinated and take sick days, inhibits their access to healthcare (“Health Equity Considerations and Racial and Ethnic Minority Groups,” 2021).

***Young Adults***

On the other hand, growing vaccine scepticism among young adults could possibly impede the U.S government’s goal of reaching herd immunity as young adults, compared to other age groups, are more likely to transmit the virus to their peers. According to a study conducted by the University of California in 2021, one in four unvaccinated adults, ages 18 to 25, said that they “probably will not” or “definitely will not” get the COVID-19 vaccine, citing concerns over possible side effects and concerns over the safety and credibility of the vaccines, while some said that they may get inoculated at a later date after seeing if the vaccines are safe. (University of California - San Francisco, 2021). Furthermore, although the study did not offer specifications about the possible side-effects the young adult population was more concerned about, a dean at Quinnipiac University cited that one of the concerns they hear from young adults were regarding issues about possible blood clots and fertility issues in women (Deliso, 2021).

***Republicans***

Many republicans’ opposition to getting the COVID-19 vaccine is often linked to former president Trump, who repeatedly downplayed the severity of the virus by claiming that “This is a flu. This is like a flu” (Diamond, 2021a). A sentiment that experts say contributed to the current mindset of many republicans that the COVID-19 pandemic is overblown and exaggerated, hence their reluctance to get the vaccine --- which they believe may cause more harm than the virus itself because of its unknown possible long-term effects (Diamond, 2021a). In addition to these, a focus group discussion among vaccine hesitant republicans reveal that they do not necessarily consider themselves “anti-vaxers”, instead, they cite that the heavy politicization of the U.S government’s response to the pandemic solidified their distrust of the government and hence, its vaccination efforts; even mentioning that the governments political appeals to get vaccinated only cement their opposition to the vaccine (Diamond, 2021b).

***Evangelical Christians***

Evangelical Christians are citing the same reasons: distrust in the government and the safety of the vaccines, skepticism over the supposed severity of the virus, misinformation, and ignorance over how vaccines work, while others simply believe that getting vaccinated goes against their faith as they believe the vaccines to be ‘unclean’ and ‘detrimental to health’ (Lovett, 2021; Reeve & G, 2021).

### Research Problem 

It is clear that the reasons behind vaccine hesitancy cited above are *intersectional.* This means that some characteristics may be a cause of, or caused by, other characteristics; however, it is necessary for us to identify the primary reason as to why individuals are not getting vaccinated such that actions are *targetive, nuanced, and effective.* Thus, the research problem we seek to answer is determining **which demographic characteristics most correspond with low vaccination rates**, such that we are able to give policy recommendations as to how the government may reach out to these demographics as we strive toward herd immunity.

# II: Dataset, Data Wrangling, and Variable Generation
### Dataset
The dataset used in the project was formed from four different online databases. We sourced vaccination rates (% of the population that is fully vaccinated) on the state level from OurWorldinData, which was in turn sourced from the Centers for Disease Control and Prevention in the United States. In order to make the data project as timely as possible while ascertaining that the researchers are able to satisfactorily treat and interpret it, the data on vaccination rates are sourced from December 30, 2021. As for our data on characteristics, we sourced data on religiosity from the percentage of the population that is highly religious by state, collected by Statista in 2017. The percentage of the population that is non-white, by state, is sourced from the Kaiser Family Foundation (KFF) and was sourced in 2019. The percentage of the population that is aged under 25, by state, is likewise sourced from the KFF and has a timeframe of 2019. Finally, the percentage of the population that is Republican-leaning was sourced from the Pew Research Forum; it is part of its 2014 Landscape Study of beliefs in the United States. These variables on religiosity, ethnicity, age, and political leaning were selected given the abundance of literature that points to these being the primary reasons behind vaccine hesitancy, as detailed above in the introduction; these timeframes of 2014, 2017, and 2019 were selected because this was the most recent data available on the characteristics.

Each database provided a .csv file containing by-state data; these .csv files were wrangled together using BigQuery and Python to create a single dataset for use in the project.

#### Reading and joining data into usable dataframes

```python
#loading the filepaths
nonwhite = pd.read_csv(r'nonwhite.csv', index_col='State')
religious = pd.read_csv(r'religion.csv',index_col ='State')
propage = pd.read_csv(r'propage.csv',index_col='State')
republican = pd.read_csv(r'republican.csv', index_col='State')

#joining the different csv files
final_table = pd.merge(nonwhite, religious, on='State', how='left')
final_table = pd.merge(final_table, propage, on='State', how='left')
final_table = pd.merge(final_table, republican, on='State', how='left')

#dropping all unnamed columns 
final_table.drop(list(final_table.filter(regex = 'Unnamed')), axis = 1, inplace = True) 

#dropping all rows with non-state status 
final_table.drop(['United States', 'Puerto Rico', 'District of Columbia'], inplace=True) #dropping non-state rows 
print(final_table)

#uncomment the next line if the csv file is needed
#final_table.to_csv('final_data.csv', index=False) 
```

# III: Descriptive Statistics
We use descriptive statistics to obtain an answer to our objective of determining which demographic characteristics most correspond with low vaccination rates. We will employ bar graphs to show *average* rates of fully vaccinated population across the categorical groups in each categorical variable we have generated prior (e.g. comparing the average rate of a fully vaccinated population between the religious categories). We would also like to note the following:
- We use *average rates* because it does not make mathematical sense to cumulate the vaccination rate data we have as these are individual rates for each state. 
- Moreover, we do not use counts of doses distributed or total number of people by state because such data may skew the results; for example, very religious states may have a very low number of vaccinated people in each state, *but* due to the number of religious states and their large populations respectively, it may *appear* that highly religious states have as many vaccinated people as slightly religious states. This kind of result would be skewed as it does not take into account the *proportion of people* fully vaccinated in each state. **Given the nature of herd immunity as requiring a large *proportion* of people vaccinated as opposed to a sheer number, the proportion or rate of those fully vaccinated across demographics is a more important consideration.**
- Finally, we used quintiles to "chop up" the dataset in order to create our five categories of intensity for each characteristic. This is because quintiles sort out data into fifths based on their data values for each characteristic, where the lowest fifth has the lowest 20% of the whole data values, and the highest fifth has the highest 20%. The approach is thus much more statistical than for the researchers to set categories themselves.

These descriptive statistics form no clear trend. This defeats the expectations formed in our introduction, wherein we expect that non-white individuals, typically in more demanding occupations and from poorer socioeconomic backgrounds, find it more difficult to be vaccinated as this comes at an opportunity cost. This lack of support for expectations may be because other characteristics find intersection at the extremes of white and non-white ethnicity. The first of these is in political leaning: Hajnal and Horowitz (2014) and Prysby (2017) found that whiter communities tend to be more Republican, with the American National Election study discovering that ⅔ of white Americans believing that Republican policies favored white demographics more and Democratic policies favored minority demographics more (Prysby, 2017). The reverse is true for voters from racial minorities, with African American and Hispanic communities almost unified in their support for the Democratic party (White & Laird, 2020; Igielnik & Budiman, 2020). And as explored in earlier sections, vaccination and other health measures are typically promoted by the Democratic Party and its supporters; this may thus be the reason for high vaccination rates for the very non-white and predominantly non-white categories which have more members of racial minorities in these states. Another evidence of intersectionality is the relationship between race and socioeconomic status; white Americans typically have higher socioeconomic status than their non-white counterparts and are more likely to be employed in white-collar jobs (Williams, Priest, & Anderson, 2016). They thus have a lower opportunity cost when it comes to taking leave in order to be vaccinated. 

Overall, it may be necessary to use other methods (e.g. regression) to determine the true relationship between race and vaccination as descriptive statistics are limited.

# IV: Pearson's R
### Reasoning
Why use Pearson's R? It is simply because Pearson's R can indicate whether a correlational relationship exists between two variables, and whether this relationship is negative (an inverse relationship) or positive (a direct relationship). This allows us to investigate the relationship between the characteristics and vaccination rates in a more statistical manner than descriptive statistics alone. Moreover, Pearson's R can produce *p-values*, which can indicate to use whether or not the correlational relationship discovered is statistically significant. A p-value less than 0.05 would mean that the correlational relationship found by Pearson's R is statistically significant, enabling us to reject the null hypothesis and confirm our a-priori expectations. A p-value of more than 0.05, however, would mean that we would retain the null hypothesis. 

### Interpretation of R-Values: 
* If = 1: perfect positive linear relationship
* If > 0: positive correlation
* If = 0: no linear relationship
* If < 0: negative correlation
* If = -1: perfect negative linear relationship

### A-Priori Expectations and Hypothesis

We first detail our a-priori expectations of the relationship between each of the independent variables and the dependent variable (by-state rate of fully vaccinated population). These will serve as the hypotheses and will be confirmed or rejected by Pearson's R. A negative relationship between an independent variable and dependent variable would indicate an inverse relationship between the two, wherein an increase in the dependent variable is associated with a decrease in the rate of the population that is fully vaccinated. A positive relationship would indicate the reverse. Our a-priori expectations are the following:

* *Relationship of % non-white population.* We expect it to be an inverse 
relationship. That is, an increase in the percentage of the population that is non-white will be *associated with a decrease* in the percentage of the population that is fully vaccinated. We form this a-priori expectation as it is the scarcely non-white population who had the highest vaccination rate in our descriptive statistics; moreover, literature suggests that non-white individuals are less likely to be fully vaccinated given the high opportunity cost of seeking out vaccination as a result of minorities’ usually lower socioeconomic status in America. 

* *Relationship of % of Republican-leaning population.* We expect it to be an inverse relationship. That is, an increase in the percentage of the population that is Republican-leaning will be *associated with a decrease* in the percentage of the population that is fully vaccinated. We form this a-priori expectation because of two reasons: first, the descriptive statistics clearly show an inverse relationship between the two variables; second, an abundance of literature (detailed in earlier sections) strongly suggests that the Republican Party and its supporters oppose both vaccination and health restrictions on account of Republican nationalist values. 

* *Relationship of religiosity.* We expect it to be an inverse relationship. That is, an increase in the percentage of the highly religious population will be *associated with a decrease* in the percentage of the population that is fully vaccinated. We form this a-priori expectation because of two reasons: the first is our descriptive statistics, which clearly show an inverse relationship between the two variables; second, an abundance of literature (detailed in earlier sections) strongly suggests that highly religious individuals in America oppose vaccination and health restrictions due to religious values of God-given liberty and freedom, as well as historical skepticism over scientific ventures.

* *Relationship of % of population aged less than 25.* We expect it to be an inverse relationship. That is, an increase in the percentage of the population below the age of 25 will be *associated with a decrease* in the percentage of the population that is fully vaccinated. We form this a-priori expectation ebcause of the trend found in the descriptive statistics, and because the literature suggests that younger individuals are more skeptical of the vaccine; this is because of fears over their fertility, the "untestedness" of vaccines, and assurance in their own immunity.

* The null hypothesis is the reverse of each of the above, wherein the relationship(s) and/or model are insignificant in explaining the relationship between the dependent and independent variables. 

The researchers will test these hypotheses using the 95% confidence bound; each hypothesis will be confirmed if the relationships and the model have a p-value of less than 0.05, indicating that the researchers can reject the null hypothesis at a 95% confidence level.
